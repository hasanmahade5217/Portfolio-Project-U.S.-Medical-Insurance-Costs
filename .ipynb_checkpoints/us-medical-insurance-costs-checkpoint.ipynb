# U.S. Medical Insurance Costs

#### Project Scoping: 
Before strat working on it's necessary to determine the features of this project based on the goals we want to achieve. After carefully inspecting the U.S. medical insurance cost dataset I have decided the following features. 

Overall analysis:
+ No of patients.
+ Find the average age of the patients. 
+ No. of Male and Female patients.
+ Find the geographical location of the patients.
+ No. of smoker vs Non Smoker
+ Average Cost of insurance.   
+ Average cost of insurance for smoker vs non-smoker and (male and female)
+ Finding out how many children a patients has. 

Areawise analysis:
+ No. of patients in the area.
+ Average age of the patients. 
+ How many of them are male and female?
+ Average age of male and female patients. 
+ Number of smoker and Non-smoker in (male and female) patients. 
+ Average age of the smokers in (male and female) patients. 
+ Average number of children (Smoker vs Non Smoker) and (male and female)
+ Average cost of insurance.
+ Average cost of insurance for smoker vs non-smoker and (male and female)

    

```python
# necessary libraries
import csv
from collections import Counter
import textwrap
```

#### Importing Dataset.

```python
# This function will read the dataset and return it in a form of dictionary.

def import_dataset(file_path):
    columns = {}
    
    with open(file_path, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        
        # Get column names
        column_names = reader.fieldnames
        
        # Initialize lists for each column
        for column in column_names:
            columns[column] = []
        
        # Store values in lists
        for row in reader:
            for column in column_names:
                columns[column].append(row[column])
    
    return columns
```

```python
file_path = 'insurance.csv'
data = import_dataset(file_path)

# Accessing different columns of the dataset
print(data['age'])  # Print values in the 'age' column
print(data['charges'])  # Print values in the 'charges' column
```

#### Overall Analysis 

```python
class OverallAnalysis:
    #default constructor
    def __init__(self, data):
        self.data = data
        

    def patient_count(self):
        total_patient = len(self.data['age'])
        
        return total_patient
    
    def average_age(self):
        sum_age = 0
        for age in self.data['age']:
            sum_age += int(age)
        
        avg_age = sum_age/len(self.data['age'])

        return avg_age
    
    def male_and_female_count(self):
        total_male = 0
        total_female = 0

        for gender in self.data['sex']:
            if gender == 'male':
                total_male += 1
            else:
                total_female += 1

        return total_male, total_female   

    def geographical_locations(self):
        location = []
        
        for area in data['region']:
            if area not in location:
                location.append(area)

        return location
    
    def somker_and_nonsmoker_count(self):
        total_smoker = 0
        total_nonsmoker = 0

        for smoking_status in self.data['smoker']:
            if smoking_status == 'yes':
                total_smoker += 1
            else:
                total_nonsmoker += 1
        
        return total_smoker, total_nonsmoker
    
    def male_and_female_smoker_count(self):
        male_smoker = 0
        female_smoker = 0

        for index in range(len(self.data['smoker'])):
            if self.data['smoker'][index] == 'yes':
                if self.data['sex'][index] == 'male' :
                    male_smoker += 1
                elif self.data['sex'][index] == 'female': 
                    female_smoker += 1

        return male_smoker, female_smoker
    
    
    def average_insurance_cost(self):
        sum_cost = 0
        for cost in self.data['charges']:
            sum_cost += float(cost)
        
        average_cost = sum_cost/len(self.data['charges'])

        return average_cost
    
    def average_male_smokers_and_nonsmokers_insurance_cost(self):
        sum_cost_smoker = 0
        sum_cost_nonsmoker = 0

        for index in range(len(self.data['sex'])):
            if self.data['sex'][index] == 'male' and self.data['smoker'][index] == 'yes':
                sum_cost_smoker += float(self.data['charges'][index])
            elif self.data['sex'][index] == 'male' and self.data['smoker'][index] == 'no':
                sum_cost_nonsmoker += float(self.data['charges'][index])

        average_cost_for_smoker = sum_cost_smoker/len(self.data['charges'])
        average_cost_for_non_smoker = sum_cost_nonsmoker/len(self.data['charges'])


        return average_cost_for_smoker, average_cost_for_non_smoker
    
    def average_female_smokers_and_nonsmokers_insurance_cost(self):
        sum_cost_smoker = 0
        sum_cost_nonsmoker = 0

        for index in range(len(self.data['sex'])):
            if self.data['sex'][index] == 'female' and self.data['smoker'][index] == 'yes':
                sum_cost_smoker += float(self.data['charges'][index])
            elif self.data['sex'][index] == 'female' and self.data['smoker'][index] == 'no':
                sum_cost_nonsmoker += float(self.data['charges'][index])

        average_cost_for_smoker = sum_cost_smoker/len(self.data['charges'])
        average_cost_for_non_smoker = sum_cost_nonsmoker/len(self.data['charges'])


        return average_cost_for_smoker, average_cost_for_non_smoker
    
    def no_of_childrens_for_female(self):
        
        smoker_children =[]
        non_smoker_children = []

        for index in range(len(self.data['sex'])):
            if self.data['sex'][index] == 'female':
                if self.data['smoker'][index] == 'yes':
                    smoker_children.append(self.data['children'][index])
                else:
                    non_smoker_children.append(self.data['children'][index])

         # Count occurrences of each unique value
        counted_childrens_smokers = Counter(smoker_children)
        counted_childrens_nonsmokers = Counter(non_smoker_children)   

        return counted_childrens_smokers, counted_childrens_nonsmokers

    

```

```python
#object creation
overall_patients_info = OverallAnalysis(data)

# Total number of patients
no_of_patients = overall_patients_info.patient_count()
print("Total number of patients: {}".format(no_of_patients))

# Total male and female patient
male_patient_count, female_patient_count = overall_patients_info.male_and_female_count()
print("Male patients: {}".format(male_patient_count))
print("Female patients: {}".format(female_patient_count))

# Average age of the patients
avg_patient_age = round(overall_patients_info.average_age())
print("Average age of the patients: {}".format(avg_patient_age))

# Patients location
patiens_geographical_locations = overall_patients_info.geographical_locations()
patiens_geographical_locations_str = ', '.join(location for location in patiens_geographical_locations)

print("Patients are generally recorded from {} regions.".format(patiens_geographical_locations_str))

# Patients: smokers vs non-smokers
smoker_count, non_smoker_count = overall_patients_info.somker_and_nonsmoker_count()
print("Smoker: {} ".format(smoker_count))
print("Non-Smoker: {} ".format(non_smoker_count))

# Patient: Male Smoker vs Female Smoker
male_smoker_count, female_smoker_count = overall_patients_info.male_and_female_smoker_count()
print("Male Smoker: {} ".format(male_smoker_count))
print("Female Smoker: {} ".format(female_smoker_count))

# Average cost of the insurance
avg_insurance_cost = overall_patients_info.average_insurance_cost()
print("Average insurance cost: {}".format(avg_insurance_cost)) 

# Insurance cost of male smoker vs male non-smoker
avg_male_smoker_insurance_cost, avg_male_nonsmoker_insurance_cost = overall_patients_info.average_male_smokers_and_nonsmokers_insurance_cost()
# Insurance cost of female smoker vs female non-smoker
avg_female_smoker_insurance_cost, avg_female_nonsmoker_insurance_cost = overall_patients_info.average_female_smokers_and_nonsmokers_insurance_cost()

print("Average male smoker insurance cost: ${:.2f}".format(avg_male_smoker_insurance_cost))
print("Average male non-smoker insurance cost: ${:.2f}".format(avg_male_nonsmoker_insurance_cost))

print("Average female smoker insurance cost: ${:.2f}".format(avg_female_smoker_insurance_cost))
print("Average female non-smoker insurance cost: ${:.2f}".format(avg_female_nonsmoker_insurance_cost))

# Children distribution: Female smoker vs non-smoker
female_smokers_child, female_non_smokers_child  =  overall_patients_info.no_of_childrens_for_female()

print("Female smokers children distribution:")
print(female_smokers_child)

print("Female non-smokers children distribution:")
print(female_non_smokers_child)


```
